```python
IMG_ROOT = '/home/quyet-invisible/quyetnt/data_sda1/data/'
```

```python
img_dirs = [os.path.join(IMG_ROOT, f'{x}') for x in os.listdir(IMG_ROOT)]
```

```python
len(img_dirs)
```

```python
from tqdm import tqdm
import shutil
```

```python
softmax = torch.nn.Softmax(dim=1)
for dir in tqdm(img_dirs):
    img = cv2.imread(dir)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_transform = transform(img)
    img_transform = img_transform.unsqueeze(0)
    pred = model(img_transform.to('cuda:0'))
    pred = softmax(pred)
    acc = torch.max(pred, axis=1)[0]
    label = torch.argmax(pred, axis=1)
    if(acc > 0.5):
        print(f'Image: {dir} - PRED: {NUM2LABEL[label.item()].upper()} - Accuracy: {acc.item()}')
        try:
            dest = ROOT_DIR + 'aicycle_data' + f'/{NUM2LABEL[label.item()]}'
            print(dest)
            shutil.copy2(dir, dest)
        except Exception as e:
            print(f'An error occur at {dir} - {e}')
    else:
        print(f'Image: {dir} is not available. - Accuracy: {acc.item()}')
```
